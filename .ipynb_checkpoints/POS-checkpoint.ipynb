```python
import spacy
nlp=spacy.load("en_core_web_sm")
text="Elon flew to mars yesterday. He carried birayni masala with him"
doc=nlp(text)
for token in doc:
    print(token,"|", token.pos_,"|", spacy.explain(token.pos_))
```

```python
doc=nlp("Elon flew to mars yesterday. He carried birayni masala with him")
for token in doc:
    print(token,"|", token.tag_,"|", spacy.explain(token.tag_))
```

```python

text="Inflation rose again in April, continuing a climb that has pushed consumers to the brink and is threatening the economic expansion, the Bureau of Labor Statistics reported Wednesday.\n\nThe consumer price index, a broad-based measure of prices for goods and services, increased 8.3% from a year ago, higher than the Dow Jones estimate for an 8.1% gain. That represented a slight ease from Marchâ€™s peak but was still close to the highest level since the summer of 1982.\n\nRemoving volatile food and ene"
doc=nlp(text)
num_token=[]
noun_token=[]

for token in doc:
    if token.pos_=="NOUN":
        noun_token.append(token)
    elif token.pos_=="NUM":
        num_token.append(token)
```

```python
count=doc.count_by(spacy.attrs.POS)
count
```

```python
for k,v in count.items():
    print(doc.vocab[k].text,"|",v)
```

```python


import nltk
nltk.download('averaged_perceptron_tagger')  # Correct resource name without '_eng'
nltk.download('punkt')  # Also download punkt tokenizer which is often needed

text = ["The", "dog", "runs", "quickly"]
tags = nltk.pos_tag(text)

# Correct unpacking of tuples in the for loop
for token, tag in tags:
    print(token, "|", tag)
```
