```python
import spacy

nlp=spacy.load("en_core_web_sm")

text= """Look for data to help you address the question. Governments are good sources...http://www.data.gov/, http://www.science.gov/, http://data.gov.uk/"""
doc=nlp(text)
for token in doc:
    if token.like_url:
        print(token)
    
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")
transactions = "Tony gave two $ to Peter, Bruce gave 500 € to Steve"
doc=nlp(transactions)
for token in doc:
    if token.is_currency:
        print("currency=", token)
        if token.i > 0:
            print("Amount:", doc[token.i - 1].text)
        
        # next token (receiver after 'to')
        if token.i + 2 < len(doc):
            print("Receiver:", doc[token.i + 2].text)
        
```

```python
doc=nlp("captain america ate appple of worth 5$")
for token in doc:
    print(token, "|", token.pos_, "|", token.lemma)
```

```python
doc=nlp("captain america ate appple of worth 5$")
for ent in doc.ents:
    print(ent.text,"|", ent.label_,"|", spacy.explain(ent.label_))
```
