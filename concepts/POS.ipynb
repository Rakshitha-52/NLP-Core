```python
import spacy
nlp=spacy.load("en_core_web_sm")
text="Elon flew to mars yesterday. He carried birayni masala with him"
doc=nlp(text)
for token in doc:
    print(token,"|", token.pos_,"|", spacy.explain(token.pos_))
```

```python
doc=nlp("Elon flew to mars yesterday. He carried birayni masala with him")
for token in doc:
    print(token,"|", token.tag_,"|", spacy.explain(token.tag_))
```
